```python
from google.colab import files
import zipfile
import io
import pandas as pd
import matplotlib.pyplot as plt

# 파일 업로드
uploaded = files.upload()

# 실제 파일 이름 확인
filename = list(uploaded.keys())[0]

# 압축 해제
with zipfile.ZipFile(io.BytesIO(uploaded[filename]), 'r') as zip_ref:
    zip_ref.extractall('/content')

# 데이터 불러오기
raw_data = pd.read_csv('/content/survey_results_public.csv')

# 데이터 전처리
target_data = raw_data[['Age', 'Country', 'LanguageHaveWorkedWith', 'LearnCode']]
methods = target_data['LearnCode'].dropna().str.split(';')
exploded_methods = methods.explode()

# 시각화
size_by_methods = exploded_methods.value_counts()
plt.figure(figsize=(6,6))
size_by_methods.nlargest(10).plot.pie(autopct='%1.0f%%', startangle=140)
plt.title('Top 10 Ways Developers Learned to Code')
plt.ylabel('')
plt.tight_layout()
plt.show()
```
